```python
# imports 

from webdriver_manager.chrome import ChromeDriverManager 
import pandas as pd
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager 
driver = webdriver.Chrome(ChromeDriverManager().install())
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
```

```python
# fetches the first page to be scraped

page_url = "https://ludopedia.com.br/ranking"
driver.get(page_url)
```

```python
# get all the elements with the CSS selector "a"

Bgname = driver.find_elements(By.CSS_SELECTOR, 'a')
```

```python
# create a list to find the name and the link to the Board Game. Added some filters to make the list easier to manipulate later
BG = []

for apelido in Bgname:
    bg_url = apelido.get_attribute('href')
    Nomebg = apelido.text
    if Nomebg != ' ' and Nomebg != '' and Nomebg!= '-' and Nomebg != 'Ver':
        BG.append({'Nome Jogo': Nomebg, 'Link': bg_url})
```

```python
# fetches the second page
page_url = "https://ludopedia.com.br/ranking?pagina=2"
driver.get(page_url)
```

```python
Bgname2 = driver.find_elements(By.CSS_SELECTOR, 'a')
```

```python
# same iteration to get the elements for the second page
for apelido in Bgname2:
    bg_url = apelido.get_attribute('href')
    Nomebg = apelido.text
    if Nomebg != ' ' and Nomebg != '' and Nomebg!= '-' and Nomebg != 'Ver':
        BG.append({'Nome Jogo': Nomebg, 'Link': bg_url})
```

```python
BG
```

```python
# transforms into a dataframe
dfBG = pd.DataFrame(BG)
```

```python
# export to .csv to manipulate the data on SQL
dfBG.to_csv('dfbg1.csv')
```
